```python
import pandas as pd

csv_path = "../data/processed/tweets_sentimentos.csv"
excel_path = "../data/processed/tweets_revisao.xlsx"
df = pd.read_csv(csv_path)

df_revisao_prioritaria = df[df['confianca'] < 0.80]

with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
    # Aba 1: Todos os dados
    df.to_excel(writer, sheet_name='Todos_os_Tweets', index=False)
    df_revisao_prioritaria.to_excel(writer, sheet_name='Revisao_Prioritaria', index=False)

print(f"Arquivo salvo com sucesso em: {excel_path}")
print(f"Total de tweets: {len(df)}")
print(f"Tweets para revis√£o priorit√°ria (<90%): {len(df_revisao_prioritaria)}")
```

```python
import pandas as pd

caminho_excel = "../data/processed/tweets_revisao.xlsx"
caminho_csv_final = "../data/processed/tweets_analise_final.csv"

df_todos = pd.read_excel(caminho_excel, sheet_name='Todos_os_Tweets')
df_revisados = pd.read_excel(caminho_excel, sheet_name='Revisao_Prioritaria')

mapeamento_correcoes = dict(zip(df_revisados['texto_limpo'], df_revisados['sentimento_final']))
df_todos['sentimento_final'] = df_todos['texto_limpo'].map(mapeamento_correcoes).fillna(df_todos['sentimento_final'])
df_todos.to_csv(caminho_csv_final, index=False, encoding='utf-8-sig')

print("Dataset atualizado com suas revis√µes manuais!")
print(f"Salvo em: {caminho_csv_final}")
print("\nNova distribui√ß√£o de sentimentos:")
print(df_todos['sentimento_final'].value_counts())
```
